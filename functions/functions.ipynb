##### create function to add two numbers

```python
def add(x:int|float  , y : int|float):
    """
    add two numbers

    """
    z  =  x+ y
    return z
```

```python
def add1(x:int|float  , y : int|float):
    """
    add two numbers

    """
    if isinstance(x  , (int , float)) and \
    isinstance(y  , (int , float)):
        z  =  x+y
        return z
    else:
        print('data should numeric')
    
```

# use assert

```python
def add1(x:int|float  , y : int|float):
    """
    add two numbers

    """
    assert isinstance(x  , (int , float))  , "x should be numeric"
    assert  isinstance(y  ,  (int  , float)) ,"y should be numeric"
    z  =  x+y
    return z
```

```python
add1(23,5)
```

```python
add1('data' , 'science')
```

```python
add1(12 , 'science')
```

```python
add1([12,4,5], [1,2,4])
```

```python
def add_args(*args):
    add  =  0
    for i in args:
        add  =  add+i
    return  add
    
```

# create function  for bill with tax

```python
def  total_bill(*args):
    bill_before_tax =  0
    price_after_tax  =  0
    
    for product  , price  ,  quantity in  args:
        subtotal  =  price*quantity
        print(f'subtotal  of {product} --> {subtotal}')

        bill_before_tax  =  bill_before_tax +subtotal

        print(f"Bill before tax --> {bill_before_tax}")

        price_after_tax  =  bill_before_tax + bill_before_tax*0.18

    return  f"bill after tax --> {price_after_tax}"

```

```python
total_bill(('mobile' ,  25000 , 2))
```

```python
total_bill(('mobile' ,  25000 , 2) , ('Laptop' , 65000 , 3))
```
