```python
@dataclass
class Triangle:
    a : int|float
    b : int|float
    c : int|float
    def Perimeter(self): 
        if (self.a+self.b>self.c) and \
          (self.b +self.c > self.a) and  \
           (self.a+ self.c > self.b):
           p = self.a+self.b+self.c
           return  p
    
    def Area(self):
        if self.Perimeter() is not None:
            s  =  self.Perimeter()/2

            Area  =  (s*(s-self.a)*(s-self.b)*(s-self.c))**1/2
            return f"Area of Triangle is {Area}"

        else:
            print(f"Sides do not form Triangle")
```

```python
class Simple_Calculator:

    def __init__(self, a, b):
        self.a = a
        self.b = b
 
    def add(self):
        if isinstance(self.a  ,(int , float)) and  isinstance(self.b  ,(int , float)):
            return self.a + self.b
 
    def subtract(self):
        if self.add() is not None:
            return self.a - self.b
 
    def multiply(self):
        if self.add() is not None:
            return self.a * self.b
 
    def divide(self):
        if self.add() is not None:
            if self.b !=0 :
                return self.a / self.b
            else:
                raise ZeroDivisionError("Denominator Should not be zero")

```

# Inheritance

```python
# Create class person and  employee
class  Person:
    def  __init__(self , name  , age):
        self.name  =  name  
        self.age=  age

    def introduce(self):
        return (f"I am {self.name}  and  i am  {self.age}  years old")
    


# create  employee class
class  employee(Person):
    def __init__(self, name  , age , company):
        super().__init__(name , age)
        self.company =  company

    
    def work(self):
        
        print(f"  {self.introduce()} , I  work for  {self.company}")
```

```python
p1  =  Person(name =  "Sunita" , age  =  23)
```

```python
e1.introduce()
```

# Multilevel  Inheritance

```python
# create class for employee , manager  , projectmanager
```

```python
class  Employee:
    def  __init__(self  , name  , emp_id ):
        self.name  = name
        self.emp_id  =  emp_id

    def get_employee_info(self):
        return f"Name  : {self.name} ,  ID  :  {self.emp_id}"


# create another class  MAnager

class Manager(Employee):
    def __init__(self , name  , emp_id  , department):
        super().__init__(name , emp_id)
        self.department  =  department

    # create method

    def  get_manager_info(self ):
        return f"{self.get_employee_info()} ,  Department  : {self.department}"


# create Project Manager class

class Project_Manager(Manager):
    def __init__(self , name  , emp_id  , department , Project):
        super().__init__(name , emp_id,department )
        self.Project  =  Project

    # create method

    def  get_Project_manager_info(self ):
        return f"{self.get_manager_info()} ,  Project : {self.Project} "




```

```python
e1  =  Employee(name = 'Surendra' , emp_id = 123)
```

```python
e1.emp_id
```

```python
e1.get_employee_info()
```

```python
e1.get_manager_info()
```

```python
m1  =  Manager(name = 'Sushmita' , emp_id = 1234 , department  = "IT")
```

```python
m1.get_manager_info()
```

```python
prm =   Project_Manager(name = 'Sushmita' , emp_id = 1234 , department  = "IT" , Project =  "Mobile  Application" )
```

```python
class User:
    def __init__(self,  email  , password):
        self.email  =  email
        self.password  =  password
      
    def   get_user_info(self) :
        return  f"user_email : {self.email}  , password :   {self.password}"
    

    # create class for admin

class Admin(User):
    def  __init__(self, email , password,  Access):
        super().__init__(email  , password)
        self.Access = Access

# create method

    def  get_Admin_info(self):
        return f"{self.get_user_info()} ,  Access : {self.Access}"


# Create class Regular_user

class Regular_User(User):
    def __init__(self , email  , password  ,  Subscription):
        super().__init__(email  ,  password  )
        self.Subscription  = Subscription



   # create method for regular user

    def Regular_User_info(self):
        return  f"{self.get_user_info()}  , Subscription  :  {self.Subscription}"
```

```python
from pydantic  import BaseModel ,  EmailStr ,  Field
from typing import List  ,  Literal 
```

```python
class Student(BaseModel):
    name  : str  =  Field(description  =  "Enter name here")
    age :  int =    Field(description = "enter age here" ,  min_value   = 0)
    Gender : Literal["Male", "Female"] = Field(description  = "Enter Gender here")
    marks  : float =  Field(description  = "Enter marks here" , gt =  0)
    email  :  EmailStr = Field(description  = " please  Enter valid email here")
    Hobbies  :  List[str]  = Field(description  = "Hobbies"  , default  =list) 
# define method info

    def  info(self):
        print(f"name :{self.name}")
        print(f"Age  :{self.age}")
        print(f"Gender : {self.Gender}" )
        print(f"Marks:{self.marks}")
        print(f"email:{self.email}")
        print(f"Hobbies : {self.Hobbies}")






    #
```
