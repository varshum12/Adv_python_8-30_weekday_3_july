# add Methods


```python
class Person:
    def __init__(self ,  Name ,  Age  , Gender  , Occupation):
        self.Name  =  Name
        self.Age  =  Age 
        self.Gender  = Gender
        self.Occupation  =  Occupation

    def Walk(self):
        print(f"My Name is {self.Name}")
        print(f"I am {self.Gender}")
        print(f"I am {self.Age} years old")
        print(f"I work as {self.Occupation}")

```

```python
p4  =  Person(Name  =  "Marshal" ,  Occupation  =  "Data Scientist" ,  Age =  27 , Gender  ="Male")
```

```python
p4
```

```python
p4.Name
```

```python
p4.Occupation
```

```python
p4.Walk
```
