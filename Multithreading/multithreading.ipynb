# create function for Hypotenus

```python
import numpy as np
```

```python
def  Hypotenus(a : int|float  , b : int|float):
    print("Hypotenus function  is started")
    time.sleep(2)
    h  =  np.sqrt(a**2+b**2)
    print(f"Hypotenus of triangle {h}")
    print("Hypotenus function is executed")
```

```python
Hypotenus(a =  3 , b =  6)
```

```python
%%time 
Hypotenus(a =  3 , b  = 5)
```

```python
%%time
Hypotenus(a = 12 , b  =  14)
Hypotenus(a =  4  , b  =  6)
```

# Create function for simple interest

```python
def  Simple_interest(P:int|float , N :int|float  ,  R : int|float):

    """
    p:principal amount
    N : number of years
    R : rate of interest
    
    """
    print("simple interest  function is started")
    time.sleep(3)
    I  =  P*N*R/100
    A  = P+I
    print(f"Simple  interest  to  be  given {I}  Amount to be return {A}")
    print("Simple interst  is Completed")
```

```python
%%time
Simple_interest(P =  200000  , N =  2  , R  =5.5)
```

```python
%%time
Hypotenus(a  =  34 ,  b =  45)
Simple_interest(P =  300000 ,  N  =  2 , R  = 4.5)
```

```python
%%time
Hypotenus( 23 ,  34)
Simple_interest(200000 ,  2 , 3.5)
```

```python
from threading  import Thread
```

```python
%%time
th1  =  Thread(target  = Hypotenus , args = ( 23 ,  34))
th2  =  Thread(target = Simple_interest , args = (200000 ,  2 , 3.5))

# start  the threads
th1.start()
th2.start()


# join the threads
th1.join()
th2.join()
```

# create function  to find square of number

```python
def  square(num):
    time.sleep(2)
    s  =   num**2
    print(f"Square of {num} is  ---> {s}")
```

```python
%%time
square(4)
```

# create function to find factorial of number

```python
def factorial(num):
    fact  =  1

    time.sleep(2)
    for i in range(1,  num+1):
        fact  =  fact *i
    print(f"Factorial of  {num} is  --> {fact }")
```

```python
%%time
factorial(34)
```

```python
l  =  [23,56, 5 ,7,8,3,2,9 ]
```

```python
for i in l:
    print(factorial(i))
```

# Create function  for multithreading

```python
%%time
def multithread(num : list):
    Threads = []
    for i in num:
        th =  Thread(target  = factorial , args  = (i,))
        th.start()
        Threads.append(th)

        # join threads

    for thread in Threads:
        thread.join()
    print(Threads)

```

```python
multithread(l)
```

# create function for file downloading

```python
def  download(url):
    file_name = url.split('/')[-1]
    print(f"{file_name}  is downloading")
    urlretrieve(url , file_name)
    print(f"{file_name} downloaded succesfully")
```

```python
download(url)
```

# create function for multithreading to download  multiple files

```python
def multidownload(urls_list :  list):
    Threads   =[]
    for i  in  urls_list:
        th  =  Thread(target  =  download  ,  args  = (i ,))
        th.start()
        Threads.append(th)

    for i  in Threads :
        i.join()
   
```
