# create class wikipedia

```python
import pandas as pd
import requests
from  bs4 import BeautifulSoup
from  warnings  import filterwarnings
filterwarnings('ignore')
class  Wikipedia:
    def __init__(self , http_url):
        self.http_url  = http_url
    
    # create  method to get page 
    def get_page(self):
        res =  requests.get(self.http_url)
        res.raise_for_status()
        return  res.content
    
    # create procedure to get soup 

    def get_soup(self):
        soup  =  BeautifulSoup(self.get_page())
        return  soup
    
    # create method to get page title
    def  page_title(self):
        soup  =  self.get_soup()
        title  =  soup.title
        title =  title.text
        return title

    # create  method to  get h1_tags
    def get_h1(self):
        soup  =  self.get_soup()
        h1_tags  =   soup.find_all(name  = 'h1' , class_="firstHeading mw-first-heading" )
        h1_tags  = h1_tags[0].text
        return  h1_tags


        # create method to get h2 tags
    def get_h2(self):
        soup  =  self.get_soup()
        h2_tags  =  soup.find_all(name  = 'h2')
        return  [i.text for  i in h2_tags]
    
    # create method to get h3_tags
    def get_h3(self):
        soup  =  self.get_soup()
        h3_tags  =  soup.find_all(name  = 'h3')
        return  [i.text for  i in h3_tags]
    

    # create method to get paras
    def get_paras(self):
        soup  =  self.get_soup()
        paras  =  soup.find_all(name  = 'p')
        for  i  in paras:
            print(i.text)
            print()

    # create method to get table
    def get_table(self):
        soup  =  self.get_soup()
        table  =  soup.find_all(name  = 'table' , class_ = 'wikitable')
        for  i in range(len(table)):
            print(pd.read_html(str(table[i])))

   # create method to get image link
    def get_image_link(self):
        soup  =  self.get_soup()
        a_tags  =  soup.find_all(name  = 'a' , class_  =  'mw-file-description')
        image_links  = [a_tags[i].get('href')for  i  in  range(len(a_tags))]
        return  image_links     
```
